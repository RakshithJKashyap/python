```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
subs = ['M3','DS','ADE','CO','SE','DMS']
neha= [90,86,85,70,80,83]
shravan = [90,95,88,93,90,96]
rakshith = [70,65,63,60,76,55]
```

```python
plt.figure()
plt.subplot(3,1,1)
plt.plot(subs,neha,'g--')
plt.title("Neha's Marks")
plt.xlabel('Subjects')
plt.ylabel('Marks')


plt.subplot(3,1,2)
plt.plot(subs,shravan,'ro')
plt.title("Shravan's Marks")
plt.xlabel('Subjects')


plt.subplot(3,1,3)
plt.plot(subs,rakshith,'bd')
plt.title("Rakshith's Marks")
plt.xlabel('Subjects')
```
