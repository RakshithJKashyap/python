```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
subs = ['M3','DS','ADE','CO','SE','DMS']
neha= [90,86,85,70,80,83]
shravan = [90,95,88,93,90,96]
rakshith = [70,65,63,60,76,55]
```

```python
plt.style.use('seaborn-poster')
plt.title('3rd SEM')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.plot(subs,neha,label = 'Neha')
plt.plot(subs,shravan,label = 'Shravan')
plt.plot(subs,rakshith,label = 'Rakshith')
plt.legend(loc = "best",shadow =True)
plt.grid()
plt.tight_layout()
plt.show()
```
