```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
m3=[89,94,54]
ds=[78,86,60]
ade=[87,84,65]
co=[93,87,54]
se=[80,88,55]
dms=[89.93,60]
m3_avg=(sum(m3)/3)
ds_avg=(sum(ds)/3)
ade_avg=(sum(ade)/3)
co_avg=(sum(co)/3)
se_avg=(sum(se)/3)
dms_avg=(sum(dms)/3)
avg=[m3_avg,ds_avg,ade_avg,co_avg,se_avg,dms_avg]
```

```python
sub_name=['M3','DS','ADE','CO','SE','DMS']
```

```python
plt.axis('equal')
plt.pie(avg,labels=sub_name,radius=1.5,autopct='%.2f%%',shadow=True,explode=[0,.05,.1,.15,.2,.25],startangle=114)
plt.show()
```
