```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
neha= [90,86,85,70,80,83]
shravan = [90,95,88,93,90,96]
rakshith = [70,65,63,60,76,55]
name=['Neha','Shravan','Rakshtih']
```

```python
avg_neha=(sum(neha)/6)
avg_shravan=(sum(shravan)/6)
avg_rakshith=(sum(rakshith)/6)
avg=[avg_neha,avg_shravan,avg_rakshith]
avg_class=sum(avg)/3
pass_marks=35
```

```python
xpos = np.arange(len(name))
plt.xkcd()
plt.xticks(xpos,name)
plt.title('Average Class Marks')
plt.xlabel('Names')
plt.ylabel('Scores')
plt.bar(xpos-.16,avg,width=.32,label='Average')
plt.bar(xpos+.16,avg_class,width=.32,label='Class Average')
plt.bar(xpos+.48,pass_marks,width=.32,label='Passing Marks')
plt.legend(loc='best')
plt.show()

```
